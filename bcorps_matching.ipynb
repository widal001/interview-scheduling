```python
from dataclasses import asdict

import pandas as pd

from cohortify.matcher import Matcher
```

```python
df_candidates = pd.read_csv("../Health Pathways Questionnaire (Responses) - Candidate Export.csv", index_col="Candidate")
df_positions = pd.read_csv("../Health Pathways Questionnaire (Responses) - Partner Export.csv", index_col="Position")
```

```python
df_candidates = df_candidates.dropna()
df_positions = df_positions.dropna()
df_candidates["Prefs"] = df_candidates["Preferences"].str.split("|")
df_positions["Prefs"] = df_positions["Preferences"].str.split("|")
```

```python
# extract candidate and position preferences
candidate_prefs = df_candidates["Prefs"].to_dict()
position_prefs = df_positions["Prefs"].to_dict()

# extract candidate and position capacities
candidate_capacity = df_candidates["Capacity"].to_dict()
position_capacity = df_positions["Capacity"].to_dict()
```

```python
position_capacity
```

```python
cohort = Matcher(
    proposer_prefs=position_prefs,
    recipient_prefs=candidate_prefs,
)
```

```python
match_result = cohort.assign_matches(
    p_capacity=position_capacity,
    r_capacity=candidate_capacity,
    p_min=1,
    r_min=1,
)
```

```python
len(match_result.matches)
```

```python
df_log = pd.DataFrame([asdict(log) for log in match_result.match_logs])
```

```python
df_matches = pd.DataFrame([{"Position": match[0], "Candidate": match[1]} for match in match_result.matches])
```

```python
df_matches.to_excel("../Matches.xlsx")
```
